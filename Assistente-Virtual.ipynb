```python
# Gravar audio e retonar como uma string
def gravarAudio():
    
    # Gravar o audio
    r = sr.Recognizer() # Criando um objeto Recognizer
    
    # Abrir o microfone e começar a gravar
    with sr.Microphone() as source:
        print('Diga algo!')
        audio = r.listen(source)
        
    # Usando reconhecedor de voz
    dados = ''
    try:
        dados = r.recognize_google(audio, language='pt-br')
        print('Você diz: '+dados.lower())
    except sr.UnknownValueError: # Usado para erros desconhecidos
        print('Por favor, repita novamente!')
    except sr.RequestError as e:
        print('Resultado de pesquisa em caso de erro de serviço '+ e)
    
    return dados.lower()
        
```

```python
# Função para pegar as respostas do assistente
def respostas(texto):
    print(texto)
    # convertendo texto em fala
    convert = gTTS(text= texto, lang='pt-br', slow=False)
    # Salvar o audio convertido para um arquivo
    convert.save('respostas_do_assistente.mp3')# Salvando o arquivo em mp3
    
    # Tocar o audio convertido
    os.system('start respostas_do_assistente.mp3') # Irá fazer com que o sistema abrar um programa de media
                                                   # para produzir o audio convertido
```

```python
# Função para chamar o assistente
def chamar_Assist(texto):
    chamar = ['olá computador','oi computador', 'acorde computador', 'como vai computador']
    texto = texto.lower() # Converte todo o texto em letras minúsculas
    # Verifica se o comando do usuário contêm uma palavra de chamar o assistente
    for frase in chamar:
        if frase in texto:
            return True
    # Retorna falso se não encontra a palavra no objeto texto.
    return False
    
    
```

```python
def saudacao(texto):
    
    # Saudações de entrada
    saudacoes_entrada = ['olá','oi','acorde']
    
    # Saudação de resposta
    saudacoes_resposta = ['olá, como posso ajudar?','oi, como posso ajudar?']
    
    # Se a entrada do usuário é uma saudação, então retorne uma resposta de saudação aleatória
    for palavra in texto.split():
        if palavra.lower() in saudacoes_entrada:
            return random.choice(saudacoes_resposta) + '.'
        
    # Se nenhuma saudação for detectada então retorne uma string vazia
    return ''
```

```python
# Função para pegar o primeiro e o segundo nome
def Nome(texto):
    lista_palavras = texto.split()
    
    for i in range(0, len(lista_palavras)):
        if i + 3 <= len(lista_palavras) - 1 and lista_palavras[i].lower() == 'quem' and lista_palavras[i + 1].lower() == 'é':
            return lista_palavras[i+2] + ' ' + lista_palavras[i+3]
```

```python
# While loop para que o nosso assistente sempre esteja nos ouvindo.
while True:
    # Gravar o audio
    texto = gravarAudio()
    resposta = '' # Essa variável irá dar uma resposta para qualquer comondo dado.
    
    # Verificar as palavras que chamam o assistente
    if(chamar_Assist(texto) == True):
        
        # Verifica as saudações feitas pelo usuário
        resposta = resposta + saudacao(texto)
        
        # Verifica se o usuário disse 'quem é'
        if('quem é' in texto):
            pessoa = Nome(texto)
            wiki = wikipedia.summary(pessoa, sentences=2) 
            resposta = resposta +' '+ wiki
            
        # Irá fazer com que o assistente responda ao seu comando com audio
        respostas(resposta)
```
